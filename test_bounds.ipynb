```python
import numpy             as np
import matplotlib.pyplot as plt
import wsgg              as w
%matplotlib inline
```

```python
P = 101325
XH2O = 0.01
XCO2 = 1.0

n = 1000
T = np.linspace(300,3000,n)

K = np.empty((n,5))
a = np.empty((n,5))

for i in range(n):
    K[i,:], a[i,:] = w.get_k_a(T[i], P, XCO2, XH2O)
    #print(T[i], np.min(a[i,:]))
    
plt.rc('font', size=14)
plt.plot(T,a[:,0])
plt.plot(T,a[:,1])
plt.plot(T,a[:,2])
plt.plot(T,a[:,3])
plt.plot(T,a[:,4])
plt.plot([2400,2400],[0,0.6])
plt.plot([500,500],[0,0.6])
plt.ylim([0,0.6])
plt.xlabel('T (K)')
plt.ylabel('a');
```
